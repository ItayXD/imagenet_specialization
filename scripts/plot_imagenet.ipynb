```python
import pickle
import matplotlib.pyplot as plt
import jax.numpy as jnp
import numpy as np
import os, os.path
```

```python
# retrieve all the dicts from SAVE_DIR
SAVE_DIR = None # same as SAVE_DIR in collect_stats.py
PLOT_SAVE_DIR = None # where to save plots

with open(os.path.join(SAVE_DIR, 'avg_train_losses.pkl'), 'rb') as f:
    avg_train_losses = pickle.load(f)

with open(os.path.join(SAVE_DIR, 'avg_val_losses.pkl'), 'rb') as f:
    avg_val_losses = pickle.load(f)

with open(os.path.join(SAVE_DIR, 'ens_train_losses.pkl'), 'rb') as f:
    ens_train_losses = pickle.load(f)

with open(os.path.join(SAVE_DIR, 'ens_val_losses.pkl'), 'rb') as f:
    ens_val_losses = pickle.load(f)

with open(os.path.join(SAVE_DIR, 'avg_train_accs.pkl'), 'rb') as f:
    avg_train_accs = pickle.load(f)

with open(os.path.join(SAVE_DIR, 'avg_val_accs.pkl'), 'rb') as f:
    avg_val_accs = pickle.load(f)

with open(os.path.join(SAVE_DIR, 'ens_train_accs.pkl'), 'rb') as f:
    ens_train_accs = pickle.load(f)

with open(os.path.join(SAVE_DIR, 'ens_val_accs.pkl'), 'rb') as f:
    ens_val_accs = pickle.load(f)

# retrieve true_val_logits from SAVE_DIR
with open(os.path.join(SAVE_DIR, 'true_val_logits.pkl'), 'rb') as f:
    true_val_logits = pickle.load(f)
```

# Average plots

```python
import matplotlib as mpl
c = np.arange(2, 9)
norm = mpl.colors.Normalize(vmin=c.min() - 5, vmax=c.max())
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.PuBuGn)
cmap.set_array([])
```

```python
import seaborn as sns
sns.set(font_scale=1.25)
sns.set_style('whitegrid')
def get_spaced_colors(cmap_name, k):
    # create the colormap
    cmap = plt.cm.get_cmap(cmap_name)
    # create an array of k-1 evenly spaced values between 0 and 1
    colors = cmap(np.linspace(0, 1, k))
    return colors
colors = get_spaced_colors('rainbow', 5)

plt.figure(figsize=(8, 6))
for i, w in enumerate(sorted(avg_train_losses.keys())):
    step_value_map = avg_train_losses[w]
    keys = jnp.array(list(step_value_map.keys()))
    values = jnp.stack(list(step_value_map.values()))
    batch_size = 128
    P = 2 ** 20
    plt.semilogx(1 + (keys * batch_size), values, color=colors[i], label=f'N={w}')
    # steps, losses = jnp.array(list(ens_t_truehat[w].keys())), ens_t_truehat[w].values()
    # plt.semilogx(steps * dilate, losses, '--', color=cmap.to_rgba(i), label=f'N={w} (ens)')
plt.xlabel('Images seen')
plt.ylabel('Training Loss')
plt.xlim([0.08*P, 7*P])
plt.legend()
plt.savefig(os.path.join(PLOT_SAVE_DIR, 'imagenet_losses.pdf'), bbox_inches="tight")
plt.show()
```

# same plot, averaged over 1024 data points

```python
import seaborn as sns
sns.set(font_scale=1.25)
sns.set_style('whitegrid')
def get_spaced_colors(cmap_name, k):
    # create the colormap
    cmap = plt.cm.get_cmap(cmap_name)
    # create an array of k-1 evenly spaced values between 0 and 1
    colors = cmap(np.linspace(0, 1, k))
    return colors
colors = get_spaced_colors('rainbow', 5)

base_w = 512

logit_map_base = true_val_logits[base_w]
baseline = {}
for k, v in logit_map_base.items():
    baseline[k] = jnp.mean(v, axis=0)

def mse(a, b):
    return np.sqrt(np.mean(np.square(np.array(a) - np.array(b))))

plt.figure(figsize=(8, 6))


final_diff = []

for i, w in enumerate(sorted(avg_train_losses.keys())[:-1]):
    logit_map = true_val_logits[w]
    step_value_map = {}
    for k, v in logit_map.items():
        step_value_map[k] = mse(jnp.mean(v, axis=0), baseline[k])
    keys = jnp.array(list(step_value_map.keys()))
    values = jnp.stack(list(step_value_map.values()))
    final_diff += [float(values[40])]
    batch_size = 128
    P = 2 ** 20
    plt.plot((keys * batch_size) / P, values,  label=f'N={w}', color=colors[i]) # color=cmap.to_rgba(i),
plt.xlim(0, 4)
plt.ylim([0, 18])
plt.xlabel('Epochs')
plt.ylabel('MSE distance from $N=512$ logits')
plt.legend()
plt.savefig("imagenet_logit_dist.pdf", bbox_inches="tight")
plt.show()
```
